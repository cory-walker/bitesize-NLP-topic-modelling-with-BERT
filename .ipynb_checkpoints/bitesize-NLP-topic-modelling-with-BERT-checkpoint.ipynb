# NLP Topic modelling with BERT
Today we will be clustering cleaned 20newsgroups comments by:
<ol>
    <li>Converting each comment into a 768 dimensional vector array using BERT</li>
    <li>Reducing each array's dimensionality via UMAP (Uniform Manifold Approximation and Projection)</li>
    <li>Clustering the UMAP arrays with hdbscan</li>
    <li>Converting each cluster into one large document, and then extracting the key words using TF-IDF (term frequencyâ€“inverse document frequency)</li>
</ol>
We learned about logging <a href=https://github.com/cory-walker/bitesize-Logging-in-Python>here</a>

```python
import logging
import pandas as pd
import datetime
import umap
import numpy as np
import hdbscan

from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.datasets import fetch_20newsgroups

# Parameters
log_file_name = 'topic-modelling-with-BERT.log'

# UMAP parameters
umap_n_neghbors = 30
umap_n_components = 5
umap_metric = 'cosine'

# HDBSCAN parameters
hdbscan_min_cluster_size = 15
hdbscan_metric='euclidean'
hdbscan_cluster_selection_method = 'eom'

# Variables
prog_start_dtme = datetime.datetime.now()
TS = datetime.datetime.now()

# Functions and procedures
def updateLogAndTs(comment):
    logging.info(comment)
    TS = datetime.datetime.now()
# Set logging
logging.basicConfig(filename=log_file_name, filemode='w', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Load raw data
logging.info('loading 20newdgroups-all',data_file_name)
df = pd.DataFrame(fetch_20newsgroups(subset='all')['data'])
```

1. Transform our comments into BERT embeddings

```python
updateLogAndTs('generating BERT embeddings')
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
bert_emb = model.encode(list(df[0]))
logging.info('generate BERT embeddings runtime: %s',datetime.datetime.now() - TS)
```

2. Reduce the BERT vectors' dimensionality using UMAP

```python
updateLogAndTs('generating UMAP embeddings')
umap_emb = umap.UMAP(n_neighbors=umap_n_neghbors, n_components=umap_n_components, metric=umap_metric).fit_transform(bert_emb)
logging.info('generate UMAP embeddings: n_neighbors: %s, n_components: %s, metric: %s, runtime: %s'
             ,umap_n_neghbors
             ,umap_n_components
             ,umap_metric
             ,datetime.datetime.now() - TS)
```

3. Cluster the documents using the UMAP vectors with HDBSCAN

```python
updateLogAndTs('generating HDBSCAN clusters')
cluster = hdbscan.HDBSCAN(min_cluster_size = hdbscan_min_cluster_size, metric = hdbscan_metric, cluster_selection_method = hdbscan_cluster_selection_method).fit(umap_emb)
logging.info('generating HDBSCAN clusters: min_cluster_size: %s, metric: %s, cluster_selection_method: %s, runtime: %s'
             ,hdbscan_min_cluster_size
             ,hdbscan_metric
             ,hdbscan_cluster_selection_method
             ,datetime.datetime.now() - TS)
```

4. a) Convert the clusters into large documents

```python
updateLogAndTs('Merging documents by topic')
df['topic'] = cluster.labels_
amalgamated_topics = df.groupby(['topic'], as_index = False).agg({0: ' '.join})
logging.info('Merging documents by topic complete. runtime: %s',datetime.datetime.now() - TS)
```

4. b) Generate TF-IDF

```python
def c_tf_idf(documents, m, ngram_range=(1, 1)):
    count = CountVectorizer(ngram_range=ngram_range, stop_words='english').fit(documents)
    t = count.transform(documents).toarray()
    w = t.sum(axis=1)
    tf = np.divide(t.T, w)
    sum_t = t.sum(axis=0)
    idf = np.log(np.divide(m, sum_t)).reshape(-1, 1)
    tf_idf = np.multiply(tf, idf)
    
    return tf_idf, count

updateLogAndTs('Generating TF-IDF')
tf_idf, count = c_tf_idf(amalgamated_topics[0].values, m=df.shape[0])
logging.info('Generating TF-IDF complete. runtime: %s',datetime.datetime.now() - TS)
```

4. c) Extract the key words for each topic

```python
def extract_top_n_words_per_topic(tf_idf, count, labels, n=20):
    words = count.get_feature_names()
    tf_idf_transposed = tf_idf.T
    indices = tf_idf_transposed.argsort()[:, -n:]
    top_n_words = {label: [(words[j], tf_idf_transposed[i][j]) for j in indices[i]][::-1] for i, label in enumerate(labels)}
    return top_n_words

top_n_words = extract_top_n_words_per_topic(tf_idf,count,list(amalgamated_topics['topic']))
topic_sizes = df.groupby(['topic'])[0].count().reset_index().rename({"topic":"topic",0:"size"}, axis="columns").sort_values("size", ascending=False)
topic_sizes['% documents'] = topic_sizes['size'] / topic_sizes['size'].sum()
topic_sizes.head() # Display the top 5 topics
```

```python
# Display the words for the top topic cluster. topic -1 means unclustered
top_n_words[topic_sizes['size'][1]]
```

Use the below code to reduce the embedding dimensionality down to two, and then plot it as a scatter plot

```python
import matplotlib.pyplot as plt

umap_data = umap.UMAP(n_neighbors=15, n_components=2, min_dist=0.0, metric='cosine').fit_transform(bert_emb)
result = pd.DataFrame(umap_data, columns=['x', 'y'])
result['labels'] = cluster.labels_

# Visualize clusters
fig, ax = plt.subplots(figsize=(10, 10))
outliers = result.loc[result.labels == -1, :]
clustered = result.loc[result.labels != -1, :]
plt.scatter(outliers.x, outliers.y, color='#BDBDBD', s=0.05)
plt.scatter(clustered.x, clustered.y, c=clustered.labels, s=0.05, cmap='hsv_r')
plt.colorbar()
plt.show()
```
